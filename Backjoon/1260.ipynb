```python
from collections import deque

N, M, V = map(int, input().split())
graph = [[0]*(N) for _ in range(N)]

for _ in range(M):
    x, y = map(int, input().split())
    graph[x-1][y-1] = 1
    graph[y-1][x-1] = 1

# dfs
def dfs(graph, p, visit):
  visit[p] = True
  print(p+1, end = ' ')
  for idx, val in enumerate(graph[p]):
    if val == 1 and visit[idx] == False:
      dfs(graph, idx, visit)

visit = [False]*N
dfs(graph, V-1, visit)
print()

# bfs
visit = [False]*N
result = []
queue = deque([V-1])
visit[V-1] = True

while queue:
    p = queue.popleft()
    print(p+1, end = ' ')

    for idx, val in enumerate(graph[p]):
        if val == 1 and visit[idx] == False:
            queue.append(idx)
            visit[idx] = True
```
